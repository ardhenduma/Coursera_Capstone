<h1>Part1<h1>

```python
import pandas as pd
```

**_Reading the table from wikipedia_**

```python
link= "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df=pd.read_html(link,header=[0])
df1=df[0]
df1.head()
```

```python
df2=df1.rename(columns={"Postal Code":"PostalCode"})
```

```python
df2.head()
```

<h3>Removing the rows that doesn't have a Borough <h3>

```python
df3=df2[df2.Borough!="Not assigned"]
df3.reset_index(drop=True,inplace=True)
```

```python
df3.head()
```

```python
df3 = df3.groupby('PostalCode',as_index=False).agg(lambda x: ','.join(set(x.dropna())))
df3.head()
```

<h2>If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough<h2>

```python
df3.loc[df3['Neighborhood'] == 'Not assigned','Neighborhood'] = df3['Borough']
df3.head()
```

<h3>Making the dataframe same as the question<h3>

```python
columns = ["PostalCode", "Borough", "Neighborhood"]
problemdf = pd.DataFrame(columns=columns)

code_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for code in code_list:
    problemdf = problemdf.append(df3[df3["PostalCode"]==code], ignore_index=True)
    
problemdf
```
